```python
import pandas as pd
```

```python
csv = pd.read_csv('classified_test_gpt.csv')
```

```python
csv
```

```python
csv['contains_ref'] = csv['contains_ref'].apply(lambda x: 1 if x == 'True' else 0)
```

```python
csv['pred_ref'] = csv['pred_ref'].apply(lambda x: 1 if (x == 'True' or x == True) else 0)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
true = csv['contains_ref'].to_list()
pred = csv['pred_ref'].to_list()
```

```python
precision_score(true, pred, average="macro", pos_label=1)
```

```python
recall_score(true, pred, average="macro", pos_label=1)
```
